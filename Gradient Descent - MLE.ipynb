```python
def mag(matrix):
    mat1 = 0
    mat2 = 0
    for i in range(len(matrix[0])):
        mat1 += matrix[0][i]**2
        mat2 += matrix[1][i]**2
    return (np.sqrt(mat1), np.sqrt(mat2))
```

```python
def GradientDescent():
    alpha = 0.01
    epsilon = 0.0001
    
    initial_vals = np.random.randn(2, len(CleanedBenignDataMatrix[0]))
    
    del_f_matrix = np.array([(np.sum((initial_vals[0, :] - CleanedBenignDataMatrix), axis=0)/initial_vals[1, :]**2),
                    ((len(CleanedBenignDataMatrix)/initial_vals[1]) -
                    (np.sum(((CleanedBenignDataMatrix - initial_vals[0, :])**2), axis=0)/initial_vals[1, :]**3))]) 
    while(True):
        final_vals = initial_vals - alpha * del_f_matrix
        
        if np.array(mag(np.abs(f(CleanedBenignDataMatrix, final_vals[0, :], final_vals[1, :]) - 
                  f(CleanedBenignDataMatrix, initial_vals[0, :], initial_vals[1, :])))).all() < epsilon:
            break
        print(initial_vals, '\n-------\n', final_vals, '\n-------------------------------------\n')
        initial_vals = final_vals
```
