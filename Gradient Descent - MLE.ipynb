```python
# Cleaning of data
BenignData = data[data['diagnosis'] == 'B']
CleanedBenignData = BenignData.drop(columns=['id', 'diagnosis', 'Unnamed: 32'], axis=1)
CleanedBenignDataMatrix = np.array(CleanedBenignData)
```

```python
def f(data, mean, sigma):
    mean.resize(1, 30)
    sigma.resize(1, 30)
    return -np.sum(np.log(sc.norm.pdf(data, mean, sigma)))
```

```python
def DifferenceFunc(data, *args):
        '''
            This function takes data, args as arguments.
            In args parameters should be initial_vals(i.e mean-initial, sigma-initial), 
            and final_vals(i.e. mean-final, sigma-final)
            This function returns the absolute difference between two function values as matrix of 30 X 1 
            or we can say that a vector of 30 vectors, of which one vector is 
            formed by magnitude of all the 357 rows of single vector/feature say 'radius_mean'
        '''
        
        # difference between two functions at instantaneous values 
        difference = f(data, args[0][0, :], args[0][1, :]) - f(data, args[1][0, :], args[1][1, :])
#         print(difference)
        # difference.shape in this example is (357, 30)
        # Converting the difference to (1 X 30) vector form 
        # by taking the square root of magnitude of vector, i.e, 30
        vector = []
        for i in range(len(difference[0])):
            temp = 0
            for j in range(len(difference)):
                temp += difference[j][i]**2
            vector.append(temp)
        vector = np.sqrt(vector)
#         print(vector)
```
