```python
import numpy as np
import pandas as pd
import scipy.stats as sc
```

```python
data = pd.read_csv('../tumor_data.csv')
```

```python
data.head()
```

```python
# Cleaning of data
BenignData = data[data['diagnosis'] == 'B']
CleanedBenignData = BenignData.drop(columns=['id', 'diagnosis', 'Unnamed: 32'], axis=1)
CleanedBenignDataMatrix = np.array(CleanedBenignData)
```

```python
def f(data, mean, sigma):
    return -(np.log(sc.norm.pdf(data, mean, sigma)))
```

```python
def DifferenceFunc(data, *args):
        '''
            This function takes data, args as arguments.
            In args parameters should be mean-initial, sigma-initial, mean-final, sigma-final
            This function returns the absolute difference between two function values as matrix of 2 X 1 
            having one agrument as mean vector and second argument as sigma vector
        '''
        
        # difference between two functions at instantaneous values 
        difference = f(data, args[0, :], args[1, :]) - f(data, args[2, :], args[3, :])
        
        # Converting the difference to mean and sigma vector form 
        # by taking the square root of magnitude of vector
        vectors = [0, 0]
        for i in range(len(difference[0])):
            vectors[0] += (difference[0][i]**2)
            vectors[1] += (difference[1][i]**2)
        vectors = np.sqrt(vectors)
        print(vectors)
```

```python
def GradientDescent():
    
    alpha = 0.01
    epsilon = 0.0001
    
    initial_vals = np.random.randn(2, len(CleanedBenignDataMatrix[0]))
    
    mu = np.sum((initial_vals[0, :] - CleanedBenignDataMatrix), axis=0)/initial_vals[1, :]**2
    
    sigma = (len(CleanedBenignDataMatrix)/initial_vals[1])-(np.sum(((CleanedBenignDataMatrix - initial_vals[0, :])**2), axis=0)/initial_vals[1, :]**3)
    
    del_f_matrix = np.array([mu, sigma]) 
    print(del_f_matrix)
    
    # TODO: Change to while and find the sol for infinite loop
    for i in range(10):
        final_vals = initial_vals - alpha * del_f_matrix
        
        if DifferenceFunc(CleanedBenignDataMatrix, del_f_matrix).all() < epsilon:
            break
        print(initial_vals, '\n-------\n', final_vals, '\n-------------------------------------\n')
        initial_vals = final_vals
```
